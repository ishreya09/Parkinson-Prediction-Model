```python
plt.figure(figsize=(30, 15))     
sns.boxplot(data=df)
```

```python
# Remove Outliers based on IQR
Q1 = df['MDVP:Fhi(Hz)'].quantile(0.25)
Q3 = df['MDVP:Fhi(Hz)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filtered_df = df[(df['MDVP:Fhi(Hz)'] >= lower_bound) & (df['MDVP:Fhi(Hz)'] <= upper_bound)]
```

```python
print("Q1 = ",Q1)
print("Q3 = ",Q3)
print("IQR = ",IQR)
print("Lower Bound = ",lower_bound)
print("Upper Bound = ",upper_bound)
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=filtered_df, x='MDVP:Fhi(Hz)')

# Add title
plt.title('Box Plot with Outliers Removed')
```
